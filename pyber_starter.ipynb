```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
combined_city_ride_df = pd.merge(ride_df, city_df, how ='left', on= 'city')
# Display the data table for preview
combined_city_ride_df.head()
```

```python
combined_city_ride_df.columns
```

```python
group_city = combined_city_ride_df.groupby('city')
group_city.head()
```

```python
u_group = combined_city_ride_df.loc[combined_city_ride_df['type'] == "Urban"]
```

```python
u_city = u_group.groupby('city')
```

```python
ucounts_city = u_city['ride_id'].count()
uaverage_fare = u_city['fare'].mean()
udrivers = u_city['driver_count'].mean()
```

```python
s_group = combined_city_ride_df.loc[combined_city_ride_df['type'] == "Suburban"]
```

```python
s_city = s_group.groupby('city')
```

```python
scounts_city = s_city['ride_id'].count()
saverage_fare = s_city['fare'].mean()
sdrivers = u_city['driver_count'].mean()
```

```python
r_group = combined_city_ride_df.loc[combined_city_ride_df['type']=="Rural"]
```

```python
r_city = r_group.groupby('city')
```

```python
rcounts_city = r_city['ride_id'].count()
raverage_fare = r_city['fare'].mean()
rdrivers = r_city['driver_count'].mean()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
city_type = ["Rural", "Suburban", "Urban"]
# Build the scatter plots for each city types
plt.scatter(ucounts_city, uaverage_fare, marker="o", facecolor = "lightcoral", edgecolor = "black", s=udrivers*10, alpha=0.60, label = "Urban")
plt.scatter(scounts_city, saverage_fare, marker="o", facecolor = "lightskyblue", edgecolor= "black", s=sdrivers*10, alpha=0.60, label = "Suburban")
plt.scatter(rcounts_city, raverage_fare, marker="o", facecolor="gold", edgecolor="black", s=rdrivers*10, alpha=0.60, label = "Rural")
plt.grid(True)

legend = plt.legend(title="City Types",loc="best",fontsize=9)
for l in legend.legendHandles:
    l.set_sizes([30.0])
plt.title("Pyber Ride Sharing Data (2016)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Average Number of Rides")
plt.text(45, 40, 'Note: Circle size correlates with driver count per city')
plt.savefig("scatterrideshare.png")
plt.show()
```

## Total Fares by City Type

```python
group_type = combined_city_ride_df.groupby('type')
city_type = ["Rural", "Suburban", "Urban"]
total_fare = group_type["fare"].sum()
explode = (0,0,0.1)
colors = ["gold", "lightskyblue", "lightcoral"]
plt.pie(total_fare, labels=city_type, explode = explode, autopct="%1.1f%%", shadow=True, startangle=150, colors=colors)
plt.axis("equal")
plt.title("% of Total Fares by City Type")
plt.savefig("piefare.png")
plt.show()
```

## Total Rides by City Type

```python
group_type = combined_city_ride_df.groupby('type')
city_type = ["Rural", "Suburban", "Urban"]
total_ride = group_type["ride_id"].count()
explode = (0,0,0.1)
colors = ["gold", "lightskyblue", "lightcoral"]
plt.pie(total_ride, labels=city_type, explode = explode, autopct="%1.1f%%", shadow=True, startangle=150, colors=colors)
plt.axis("equal")
plt.title("% of Total Rides by City Type")
plt.savefig("pieride.png")
plt.show()
```
